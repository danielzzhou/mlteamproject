```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("combine_.csv",index_col=0)
df.head()
```

```python
#df = pd.read_csv('dailyupdated.csv')
df = df.dropna(subset=['RETX'])
df=df[df['RETX']!='C']
df=df[df['RETX']!='B']
df['RETX']=df['RETX'].astype(float)
df = df[['date','CUSIP','RETX','PRC']]
df['PRC'] = df['PRC'].abs()
```

```python
df1 = pd.read_csv('spdaily.csv')
df1 = df1[['caldt','spindx','sprtrn']]
df1.head()
```

```python
new = df.join(df1.set_index('caldt'), on='date')
```

```python
grouped = new.groupby(['CUSIP'])
new = new.sort_values(by = 'date', ascending = True)
```

```python
y_raw = pd.read_csv("y.csv")
dates = y_raw["date"].unique()
dates = list(dates)
dates.sort()
dates = np.array(dates)
```

```python
K = []
for i in [10,21]:
    K.append(i)
for i in range(42,253,21):
    K.append(i)
```

```python
for k in K:
    print(k, "#############################")
    result = []
    count = 0
    a = np.array([1]*(k))
    for i, j in grouped:
        count += 1
        if count %100 == 0:
            print(count)
        #if count == 3:
        #    break
        list1 = []
        list2 = []
        idx = (dates < j["date"].iloc[0]).sum() 
        idx2 = (dates > j["date"].iloc[-1]).sum()
        datetmp = dates[idx:] if idx2 == 0 else dates[idx:-idx2]
        for m in datetmp:
            q = np.where(j["date"]==m)[0]
            if (len(q) == 0) or q[0] < k:
                list1.append('nan')
                list2.append('nan')
            else:
                q = q[0]
                b = np.array(j['sprtrn'].iloc[q-k:q])
                X = np.array([a,b]).T
                Y = np.array(j['RETX'].iloc[q-k:q])
                beta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y.T)
                list1.append(beta[0])
                list2.append(beta[1])
        
        resulttmp = pd.DataFrame()
        resulttmp['date'] = datetmp
        resulttmp['CUSIP'] = i
        resulttmp['alpha'+str(k)] = list1
        resulttmp['beta'+str(k)] = list2
        result.append(resulttmp)
    
    output = pd.concat(result, axis=0)
    output.to_csv("ff" + str(k) + ".csv", index=False)
```
