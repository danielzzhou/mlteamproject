```python
import pandas as pd
import datetime
import numpy as np
```

```python
df_daily = pd.read_csv('dailyupdated.csv')
```

```python
df_daily
```

```python
df_daily = df_daily.dropna(subset=['RETX'])
```

```python
df_daily=df_daily[df_daily['RETX']!='C']
df_daily=df_daily[df_daily['RETX']!='B']
```

```python
df_daily['RETX']=df_daily['RETX'].astype(float)
```

```python
df_daily_essentials = df_daily[['date','CUSIP','RETX']]
df_daily_essentials.head()
```

```python
len(df_daily_essentials)
```

```python
len(df_daily_essentials['CUSIP'].drop_duplicates(keep="first"))
```

```python
df_daily_essentials = df_daily_essentials.sort_values(by = 'date', ascending = True)
```

```python
df_daily_essentials['CUSIP'] = df_daily_essentials['CUSIP'].astype(str)
df_daily_essentials = df_daily_essentials.sort_values(by =['CUSIP','date'])
```

```python
df_daily_essentials = df_daily_essentials.sort_values(by =['CUSIP','date'])
```

```python
df_daily_essentials.groupby(['CUSIP']).apply(lambda x: x.sort_values(by = 'date', ascending = True))
```

```python
K = []
for i in range(1,22):
    K.append(i)
for i in range(42,253,21):
    K.append(i)
len(K)
```

```python
Momentums = []
colname = []
for k in K:
    name = 'momentum'+str(k)
    series = df_daily_essentials.groupby(['CUSIP'])['RETX'].rolling(k).sum().reset_index()
    series = pd.DataFrame(series)
    Momentums.append(series['RETX'].to_list())
    colname.append(name)
```

```python
df_daily_essentials.sort('CUSIP')
```

```python
addcol = dict(zip(colname,Momentums))
for col, mom in addcol.items():
    df_daily_essentials[col]=mom
```

```python
df_daily_essentials.to_csv('dailywithmomentum.csv')
```
